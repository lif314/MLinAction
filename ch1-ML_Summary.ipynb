## 1.1 NumPy函数库基础

```python
from numpy import *
```

```python
# 随机数组 Array
random.rand(4,4)
```

```python
# 转化为矩阵  Matrix
randMat = mat(random.rand(4,4))

invRandMat = randMat.I   # .I对矩阵进行求逆运算

invRandMat
```

```python
# 矩阵乘法
randMat * invRandMat  # 结果是单位矩阵
```

```python
# 创建单位矩阵
eye(4)
```
